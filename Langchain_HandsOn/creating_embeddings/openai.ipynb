```python
embeddings_1024 = OpenAIEmbeddings(model="text-embedding-3-small", dimensions=1024)

```

```python
embeddings_1024
```

```python
text = "this is a tutoriAl on OPENAI EMBEDIDNGS"
query_result_1024 = embeddings_1024.embed_query(text)
len(query_result_1024)
```

```python
from langchain_community.document_loaders import TextLoader
loader = TextLoader("speech.txt")
data = loader.load()
data
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
chunks = text_splitter.split_documents(data)
len(chunks)
chunks[0]
#split by sentence
#from langchain_text_splitters import SentenceTransformersTokenTextSplitter
```

```python
## vector embedding and vectore storeDB
from langchain_community.vectorstores import Chroma

db = Chroma.from_documents(chunks, embeddings_1024)
```

```python
db
```

```python
retrieved_results = db.similarity_search("what is the name of the person in the speech?")
```
